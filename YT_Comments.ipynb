# **SAVE DATA AS PANDAS**

```python
# Importing necessary modules from the Google API client library for YouTube Data API
import googleapiclient.discovery
import pandas as pd

# Setting up the parameters for the YouTube Data API
api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "AIzaSyB-2SJlE-m29A8e9udk-TeWeYzSwEaVmO4"

# Building the YouTube API client
youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)

# Creating a request to retrieve comment threads for a specific video
request = youtube.commentThreads().list(
    part="snippet",
    videoId="45mnmy2JUl0",
    maxResults=100
)

# Executing the request and storing the response
response = request.execute()

# Creating an empty list to store comment data
comments = []

# Iterating through the items in the response and extracting relevant comment information
for item in response['items']:
    # Extracting information from the snippet of the top-level comment
    comment = item['snippet']['topLevelComment']['snippet']

    # Appending comment information to the list as a list of values
    comments.append([
        comment['authorDisplayName'],
        comment['publishedAt'],
        comment['updatedAt'],
        comment['likeCount'],
        comment['textDisplay']
    ])

# Creating a Pandas DataFrame from the list of comments
df = pd.DataFrame(comments, columns=['author', 'published_at', 'updated_at', 'like_count', 'text'])

# Displaying the first 20 rows of the DataFrame
df.head(20)
```
